```python
import numpy as np 
from sklearn.preprocessing import StandardScaler
```

```python
data = np.loadtxt('ccpp_train.csv',delimiter = ',')
```

```python
data.shape
```

```python
x = data[:,:-1]
y = data[:,-1]
test_data = np.loadtxt('ccpp_test.csv',delimiter = ',')

scaler  = StandardScaler()
scaler.fit(x)
x = scaler.transform(x)
x_test = scaler.transform(test_data)

x = np.append(x,np.ones((len(x),1)),axis =1 )
x_test = np.append(x_test,np.ones((len(x_test),1)),axis = 1)

```

```python
def sg(x,y,learning_rate,m):
    M,N = x.shape
    m_slope = np.zeros(N)
    for i in range(M):
        for j in range(N):
            m_slope[j] += (-2/M)*(y[i]-(m*x[i]).sum())*x[i][j]
    new_m = m-learning_rate*m_slope
    return  new_m
```

```python
def cost(x,y,m):
    N = len(x)
    cost =0 
    for  i in range(N):
        cost += (1/N)*((y[i]-np.sum(x[i,:]*m))**2)
    return cost
```

```python
def gd(x,y,learning_rate,num_iterations):
    M,N = x.shape
    m =0 
    
    for i in range(num_iterations):
        m = sg(x,y,learning_rate,m)
        print('cost',i,':',cost(x,y,m))
    return m 
```

```python
def predict(m):
    y_predict = []
    N = len(x_test)
    for i in range(N):
        y_predict.append((m*x_test[i]).sum())
    np.savetxt('ccpp_op.csv',y_predict,fmt = '%.5f')
    print(y_predict)
```

```python
learning_rate = 0.1

num_iterations = 250
m = gd(x,y,learning_rate,num_iterations)
predict(m)
```
